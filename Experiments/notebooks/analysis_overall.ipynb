```python
import json
import numpy as np
import pandas as pd
from pathlib import Path

# =========================================================
# Config
# =========================================================
REPEATS = [1, 2, 3]
# ROUND = 149
ROUND = 149
# ROUND = 29
# ROUND = 49
TASK_WHITELIST = ["semseg", "normal", "edge", "depth", "saliency"]

TASK_METRIC_MAP = {
    "semseg": "meanIoU",
    "edge": "odsF",
    "saliency": "f1",
    "depth": "rmse",
    "normal": "meanAngularErr",
}

LOWER_IS_BETTER = {
    "meanIoU": False,
    "odsF": False,
    "f1": False,
    "rmse": True,
    "meanAngularErr": True,
}

LOCAL_ALGO = "none"


# =========================================================
# Helpers
# =========================================================
def std_ddof1(x):
    x = pd.Series(x).dropna()
    return float(np.std(x, ddof=1)) if len(x) > 1 else 0.0


def get_round_deltaM(round_obj):
    dm = None
    if isinstance(round_obj, dict):
        if round_obj.get("deltaM") is not None:
            dm = round_obj.get("deltaM")
        elif isinstance(round_obj.get("serverMetrics"), dict) and round_obj["serverMetrics"].get("deltaM") is not None:
            dm = round_obj["serverMetrics"].get("deltaM")
    return float(dm) if isinstance(dm, (int, float)) else None


def build_id_to_index_map(data):
    header_clients = (data.get("header", {}) or {}).get("clients", []) or []
    if header_clients:
        return {c.get("id"): i for i, c in enumerate(header_clients) if c.get("id") is not None}

    rounds = data.get("rounds", []) or []
    if rounds and (rounds[0].get("clientMetrics") is not None):
        cms = rounds[0].get("clientMetrics") or []
        return {cm.get("id"): i for i, cm in enumerate(cms) if cm.get("id") is not None}

    return {}


def parse_file(fp, algo_name, rep):
    with open(fp, "r", encoding="utf-8") as f:
        data = json.load(f)

    id2idx = build_id_to_index_map(data)

    rows_task = []
    rows_delta_raw = []

    for r in data.get("rounds", []) or []:
        rnd = r.get("round")

        rows_delta_raw.append({
            "algo": algo_name,
            "rep": rep,
            "round": rnd,
            "deltaM_raw": get_round_deltaM(r),  # optional/debug only
        })

        cms = r.get("clientMetrics", []) or []
        for j, cm in enumerate(cms):
            cid = cm.get("id")
            cidx = id2idx.get(cid, j)

            for head in cm.get("perHead", []) or []:
                rows_task.append({
                    "algo": algo_name,
                    "rep": rep,
                    "round": rnd,
                    "client_idx": cidx,
                    "client_id_raw": cid,
                    "task": head.get("task"),
                    "testMetric": head.get("testMetric"),
                    "testMetricGlobal": head.get("testMetricGlobal"),
                })

    df_task = pd.DataFrame(rows_task).sort_values(["algo", "rep", "round", "client_idx", "task"])
    df_delta_raw = pd.DataFrame(rows_delta_raw).sort_values(["algo", "rep", "round"])

    return df_task, df_delta_raw


def deltaM_from_means(fed: dict, local: dict) -> float | None:
    tasks = [t for t in local.keys()
             if np.isfinite(local.get(t, np.nan)) and np.isfinite(fed.get(t, np.nan))]
    if not tasks:
        return None

    s = 0.0
    n = 0
    for t in tasks:
        m_local = float(local[t])
        m_fed = float(fed[t])

        if not np.isfinite(m_local) or not np.isfinite(m_fed) or abs(m_local) < 1e-12:
            continue

        metric_name = TASK_METRIC_MAP.get(t)
        sign = -1.0 if LOWER_IS_BETTER.get(metric_name, False) else 1.0
        s += sign * ((m_fed - m_local) / m_local)
        n += 1

    return (s / n) if n > 0 else None


def compute_deltaM_series(summary_df: pd.DataFrame, local_algo: str = "none") -> pd.DataFrame:
    """
    summary_df columns: algo, rep, round, task, mean
    mean must already be averaged across clients
    returns deltaM per (algo, rep, round)
    """
    wide = summary_df.pivot_table(
        index=["algo", "rep", "round"],
        columns="task",
        values="mean",
        aggfunc="first"
    )

    rows = []

    reps = sorted(summary_df["rep"].dropna().unique())
    algos = sorted(summary_df["algo"].dropna().unique())

    for rep in reps:
        try:
            local = wide.loc[(local_algo, rep)]  # index = round, columns = task
        except KeyError:
            for algo in algos:
                rep_rounds = summary_df.loc[
                    (summary_df["algo"] == algo) & (summary_df["rep"] == rep), "round"
                ].unique()
                for rnd in rep_rounds:
                    rows.append({"algo": algo, "rep": rep, "round": rnd, "deltaM": None})
            continue

        rep_slice = summary_df[summary_df["rep"] == rep][["algo", "round"]].drop_duplicates()

        for _, rr in rep_slice.iterrows():
            algo = rr["algo"]
            rnd = rr["round"]

            if algo == local_algo:
                rows.append({"algo": algo, "rep": rep, "round": rnd, "deltaM": 0.0})
                continue

            try:
                fed_row = wide.loc[(algo, rep, rnd)]
            except KeyError:
                rows.append({"algo": algo, "rep": rep, "round": rnd, "deltaM": None})
                continue

            if rnd not in local.index:
                rows.append({"algo": algo, "rep": rep, "round": rnd, "deltaM": None})
                continue

            fed_dict = fed_row.to_dict()
            local_dict = local.loc[rnd].to_dict()
            dm = deltaM_from_means(fed_dict, local_dict)

            rows.append({"algo": algo, "rep": rep, "round": rnd, "deltaM": dm})

    return pd.DataFrame(rows).sort_values(["algo", "rep", "round"])


def summarize_metric(df_task, metric_col, local_algo="none"):
    """
    Computes:
    1. per-task mean/std across repetitions
    2. deltaM mean/std across repetitions
    using the specified metric column: testMetric or testMetricGlobal
    """

    df = df_task.copy()

    # If metric only exists in final round, filter early
    if metric_col == "testMetricGlobal":
        df = df[df["round"] == ROUND]

    # per-round mean across clients
    per_round_task = (
        df.groupby(["algo", "rep", "round", "task"])[metric_col]
        .mean()
        .reset_index(name="mean")
        .sort_values(["algo", "rep", "round", "task"])
    )

    # per-repetition overall task score (average across rounds)
    # per_rep_task = (
    #     per_round_task.groupby(["algo", "rep", "task"])["mean"]
    #     .mean()
    #     .reset_index(name="task_mean")
    #     .sort_values(["algo", "rep", "task"])
    # )
    per_rep_task = (
        per_round_task.groupby(["algo", "rep", "task"])["mean"]
        .mean()
        .reset_index(name="task_mean")
        .sort_values(["algo", "rep", "task"])
    )

    # aggregate across repetitions
    task_summary = (
        per_rep_task.groupby(["algo", "task"])["task_mean"]
        .agg(mean="mean", std=std_ddof1, n="count")
        .reset_index()
        .sort_values(["algo", "task"])
    )
    task_summary["mean_std"] = task_summary.apply(
        lambda r: f"{r['mean']:.4f} ± {r['std']:.4f}", axis=1
    )

    # deltaM from same metric column
    deltaM_series = compute_deltaM_series(per_round_task, local_algo=local_algo)

    # per-repetition overall deltaM (average across rounds)
    # per_rep_deltaM = (
    #     deltaM_series.groupby(["algo", "rep"])["deltaM"]
    #     .mean()
    #     .reset_index(name="deltaM_mean")
    #     .sort_values(["algo", "rep"])
    # )
    if metric_col == "testMetricGlobal":
        deltaM_series = deltaM_series[deltaM_series["round"] == ROUND]

    per_rep_deltaM = (
        deltaM_series.groupby(["algo", "rep"])["deltaM"]
        .mean()
        .reset_index(name="deltaM_mean")
        .sort_values(["algo", "rep"])
    )

    # aggregate deltaM across repetitions
    deltaM_summary = (
        per_rep_deltaM.groupby("algo")["deltaM_mean"]
        .agg(mean="mean", std=std_ddof1, n="count")
        .reset_index()
        .sort_values("algo")
    )
    deltaM_summary["mean_std"] = deltaM_summary.apply(
        lambda r: f"{r['mean']:.4f} ± {r['std']:.4f}", axis=1
    )

    return task_summary, deltaM_summary, per_round_task, deltaM_series


# =========================================================
# Load all repetitions
# =========================================================
all_task = []
all_delta_raw = []

for algo, base in FILE_BASES.items():
    for rep in REPEATS:
        fp = ROOT / f"{base}_{rep}.json"
        if not fp.exists():
            raise FileNotFoundError(f"Missing file: {fp}")

        dt, dd = parse_file(str(fp), algo, rep)
        all_task.append(dt)
        all_delta_raw.append(dd)

df_task = pd.concat(all_task, ignore_index=True)
df_delta_raw = pd.concat(all_delta_raw, ignore_index=True)

if ROUND is not None:
    df_task = df_task[df_task["round"] <= ROUND].copy()
    df_delta_raw = df_delta_raw[df_delta_raw["round"] <= ROUND].copy()

if TASK_WHITELIST is not None:
    df_task = df_task[df_task["task"].isin(TASK_WHITELIST)].copy()


# =========================================================
# Compute LOCAL and GLOBAL summaries
# =========================================================
task_summary_local, deltaM_summary_local, per_round_local, deltaM_local_series = summarize_metric(
    df_task, metric_col="testMetric", local_algo=LOCAL_ALGO
)

task_summary_global, deltaM_summary_global, per_round_global, deltaM_global_series = summarize_metric(
    df_task, metric_col="testMetricGlobal", local_algo=LOCAL_ALGO
)


# =========================================================
# Display
# =========================================================
print("\n=== LOCAL test set: per-task overall metric (testMetric) ===")
print(task_summary_local[["algo", "task", "mean", "std", "n", "mean_std"]].to_string(index=False))

print("\n=== LOCAL test set: overall deltaM ===")
print(deltaM_summary_local[["algo", "mean", "std", "n", "mean_std"]].to_string(index=False))

print("\n=== GLOBAL test set: per-task overall metric (testMetricGlobal) ===")
print(task_summary_global[["algo", "task", "mean", "std", "n", "mean_std"]].to_string(index=False))

print("\n=== GLOBAL test set: overall deltaM ===")
print(deltaM_summary_global[["algo", "mean", "std", "n", "mean_std"]].to_string(index=False))
```
